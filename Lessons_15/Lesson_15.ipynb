```python
# matplotlib.pyplot - это инструмент для рисования на Python. Может визуализировать процесс обработки изображений с помощью tenorflow
import matplotlib.pyplot as plt
import tensorflow as tf
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
```

```python
# Прочитать исходные данные изображения
image_raw_data1 = tf.io.gfile.GFile('drive/MyDrive/01.jpg', 'rb').read()
image_raw_data2 = tf.io.gfile.GFile('drive/MyDrive/02.jpg', 'rb').read()
```

```python
# Загрузить исходные данные изображения
img1 = load_img('drive/MyDrive/01.jpg')
img2 = load_img('drive/MyDrive/02.jpg')
data1 = img_to_array(img1)
data2 = img_to_array(img2)
```

```python
# Конвертация в тензоры
t_a = tf.convert_to_tensor(data1)
t_b = tf.convert_to_tensor(data2)
```

```python
# сложение тензоров
t_c = tf.add(t_a, t_b)
print(t_c)
```

```python
print(t_c.eval)
```

```python
img = t_c
```

```python
# Используйте инструмент pyplot для визуализации полученного изображения
plt.imshow(img)
```

```python
# сохранить изображение с новым именем файла
tf.keras.utils.save_img('drive/MyDrive/03.jpg', t_c)
```
