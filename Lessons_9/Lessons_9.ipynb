1. Объединяет два Pandas датафрейма (df1 и df2) по двум столбцам (x_1, 
y_1 и x_2, y_2), так чтобы остались только:
 общие строки – создает df3 (предусмотреть наличие аргументов 
left_on и right_on для индексов «левого» и «правого» датафремов);
 остались все строки – создает df4.

```python
# Загрузить библиотеку
import pandas as pd
import numpy as np
```

```python
# Создать фрейм данных
df1 = pd.DataFrame({'x_1': ['A', 'B', 'C'] * 3,
                    'y_1': ['D', 'E', 'F'] * 3,
                    'z': np.random.randint(0, 20, 9)})

df2 = pd.DataFrame({'x_2': ['A', 'B', 'C'] * 4,
                    'y_2': ['D', 'E'] * 6,
                    'z': np.random.randint(0, 20, 12)})
```

```python
print(df1); print('\n'); print(df2)
```

```python
# Cлияние фреймов данных (по двум столбцам, внутреннее)
df_3 = pd.merge(df1, df2, 
                how='inner',
                left_on=['x_1', 'y_1'],
                right_on=['x_2', 'y_2'],
                suffixes=['_left', '_right'])
df_3
```

```python
# Cлияние фреймов данных (по двум столбцам, внешнее)
df_4 = pd.merge(df1, df2, 
                how='outer',
                left_on=['x_1', 'y_1'],
                right_on=['x_2', 'y_2'],
                suffixes=['_left', '_right'])
df_4
```

2. Находит в датафрейме df4 позиции всех отсутствующих значений 
(формирует список из «координат» таких значений) и заменяет их значением 0

```python
# Для этого можно использовать метод fillna()
df_4.fillna(0)
```

 3. Добавляет к датафрейму df4 столбец z_avg справа, в котором вычисляется арифметическое среднее (z_left + z_right)/2.

```python
df_4['z_avg'] = (df_4.z_left + df_4.z_right)/2
```
